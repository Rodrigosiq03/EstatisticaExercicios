```python
'''
EXERCÍCIO 01
'''

# Vitórias no futebol americano (população)

import pandas as pd

vitorias = pd.Series([13, 10, 12, 11, 7, 8, 6, 6,
                      10, 7, 12, 4, 6, 5, 2, 2])

v_max = vitorias.max()
v_min = vitorias.min()

amplitude = v_max - v_min
print(f'A amplitude sobre as vitorias no futebol americano é {amplitude}')

media = vitorias.mean()
print(f'A média sobre as vitorias no futebol americano é {media:.2f}')

variancia = vitorias.var(ddof=0) # ddof=0 => desvio padrão populacional
print(f'A variância sobre as vitorias no futebol americano é {variancia:.2f}')

desvio_padrao = vitorias.std()
print(f'O desvio padrão sobre as vitorias no futebol americano é {desvio_padrao:.2f}')
```

```python
'''
EXERCÍCIO 05
'''

# Carros por domicílio

dados = pd.DataFrame({'Freq (f)': [3, 15, 24, 8],
                      'Pt. médio (x)': [0, 1, 2, 3]})

dados['x*f'] = dados['Freq (f)'] * dados['Pt. médio (x)']

# Cálculo da média
## Seu código aqui!!!
media_dados = dados['Pt. médio (x)'].mean()

# Coluna valor - média
## Seu código aqui!!!
dados['x-xbarra'] = dados['Pt. médio (x)'] - media_dados

# Coluna (valor - média)^2
## Seu código aqui!!!
dados['(x-xbarra)^2'] = dados['x-xbarra']**2

# Coluna (valor - média)^2 * frequência
## Seu código aqui!!!
dados['(x-xbarra)^2*f'] = dados['(x-xbarra)^2'] * dados['Freq (f)']
display(dados) 
# Cálculo da média amostral
## Seu código aqui!!!
obs = dados['Freq (f)'].sum()
media_amostral = dados['x*f'].sum() / obs
print(f'Media amostral: {media_amostral:.1f} carros por domicílio')
# Cálculo do desvio padrão amostral
## Seu código aqui!!!
desvpadrao_amostral = (dados['(x-xbarra)^2*f'].sum() / (obs-1))**(1/2)
print(f'Desvio padrao amostral: {desvpadrao_amostral:.1f} carros por domicílio')
```

```python
# Renda Familiar 

dados = pd.DataFrame({'Classe':['$0 - $1999', '$2000 - $3999', 
                                '$4000 - $5999', '$6000 - $7999', 
                                '$8000 - $9999', '$10000 - ou+'],
                      'Freq (f)': [11, 12, 20, 10, 13, 10],
                      'Pt. médio (x)': [999.50, 2999.50, 4999.50, 
                                        6999.50, 8999.50, 10999.50]})

dados['x*f'] = dados['Freq (f)'] * dados['Pt. médio (x)']

# Cálculo da média
## Seu código aqui!!!
media_dados = dados['Pt. médio (x)'].mean()

# Coluna valor - média
## Seu código aqui!!!
dados['x-xbarra'] = dados['Pt. médio (x)'] - media_dados

# Coluna (valor - média)^2
## Seu código aqui!!!
dados['(x-xbarra)^2'] = dados['x-xbarra']**2

# Coluna (valor - média)^2 * frequência
## Seu código aqui!!!
dados['(x-xbarra)^2*f'] = dados['(x-xbarra)^2'] * dados['Freq (f)']
display(dados) 
# Cálculo da média amostral
## Seu código aqui!!!
obs = dados['Freq (f)'].sum()
media_amostral = dados['x*f'].sum() / obs
print(f'Media amostral: {media_amostral:.1f} renda familiar')
# Cálculo do desvio padrão amostral
## Seu código aqui!!!
desvpadrao_amostral = (dados['(x-xbarra)^2*f'].sum() / (obs-1))**(1/2)
print(f'Desvio padrao amostral: {desvpadrao_amostral:.1f} renda familiar')
```

```python
'''
EXERCÍCIO 06
'''

# Notas no SAT

notas_sat = pd.DataFrame({'Notas_homens': [1520, 1750, 2120, 1380,
                                           1980, 1650, 1030, 1710],
                         'Notas_mulheres': [1790, 1510, 1500, 1950,
                                            2210, 1870, 1260, 1590]})
display(notas_sat)

# Cálculo das médias, desvios padrão e coeficiente de variação
est_notas = pd.DataFrame(notas_sat.mean(axis=0, numeric_only=True), columns=['Média'])
est_notas['DesvPad'] = notas_sat.std(axis=0, ddof=0, numeric_only=True)
est_notas['CV'] = est_notas['DesvPad']/est_notas['Média']*100
display(est_notas)
```
