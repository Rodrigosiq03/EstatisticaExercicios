```python
'''
EXERCICIO 02 (Créditos da faculdade)
'''

import pandas as pd

# Carregando o dataset (Créditos da faculdade)

creditos = pd.Series([12, 14, 16, 15, 13, 14, 15,
           18, 16, 16, 12, 16, 15])

# Calculando a média
media = creditos.mean()
print(f'A média é {media:.2f}')
# Calculando a mediana
mediana = creditos.median()
print(f'A mediana é {mediana}')
#Calculando a moda
moda = list(creditos.mode())
print(f'A moda é {moda[0]}')
```

```python
'''
EXERCICIO 05
'''

# Nota Final
import numpy as np

notas = pd.DataFrame({'Notas':[85, 80, 100, 90, 93],
                      '%nota final':[0.05, 0.35, 0.2, 0.15, 0.25]})
notas.index = ['Dever de casa', 'Testes', 'Projeto', 'Apresentações', 'Prova Final']
display(notas)

media_ponderada_notas = np.average(a=notas['Notas'], weights=notas['%nota final']) # a representa os valores e weights os pesos
print(f'A média final é {media_ponderada_notas}.')

# Graus

graus = pd.DataFrame({'Graus':[4, 3, 2, 1],
                      'Créditos':[4, 6, 3, 2]})
graus.index = ['A', 'B', 'C', 'D']
display(graus)

media_ponderada_graus = np.average(a=graus['Graus'], weights=graus['Créditos']) # a representa os valores e weights os pesos
print(f'A média final é {media_ponderada_graus}.')
```

```python
'''
Exercicio 09
'''

# A

distancias = pd.Series([28, 32, 28, 30, 34]) # carro A

media = distancias.mean()
print(f'A média das distâncias do carro A é {media:.2f}')

mediana = distancias.median()
print(f'A mediana das distâncias do carro A é {mediana:.2f}')

moda = distancias.mode()
print(f'A moda das distâncias do carro A é {moda[0]}')

# A medida de tendência central mais recomendada para esse caso é a média por ter 
# o maior valor dentre as três medidas

# B

distancias = pd.Series([31, 29, 31, 29, 31]) # carro B

media = distancias.mean()
print(f'A média das distâncias do carro B é {media:.2f}')

mediana = distancias.median()
print(f'A mediana das distâncias do carro B é {mediana:.2f}')

moda = distancias.mode()
print(f'A moda das distâncias do carro B é {moda[0]}')

# A medida de tendência central mais recomendada para esse caso pode tanto ser a
# mediana, quanto a moda por ter o mesmo valor e ser maior em relação a média

# C

distancias = pd.Series([29, 32, 28, 32, 30]) # carro C

media = distancias.mean()
print(f'A média das distâncias do carro C é {media:.2f}')

mediana = distancias.median()
print(f'A mediana das distâncias do carro C é {mediana:.2f}')

moda = distancias.mode()
print(f'A moda das distâncias do carro C é {moda[0]}')

# A medida de tendência central mais recomendada para esse caso é a moda por ter 
# o maior valor dentre as três medidas
```
