```python
import pandas as pd
from matplotlib.ticker import PercentFormatter
import matplotlib.pyplot as plt
# CONJUNTO A
dados_a = pd.DataFrame({'Valores' : [44, 65, 104, 82, 88]})
dados_a.index = ['Alemanha', 'Grã-Betanha', 
               'EUA', 'Rússia', 'China']

dados_a = dados_a.sort_values(by='Valores', ascending=False)

dados_a['PorcAcum'] = dados_a['Valores'].cumsum() / dados_a['Valores'].sum()*100

cor_barra = 'steelblue'
cor_linha = 'red'
tam_linha = 4

fig_a, pareto_a = plt.subplots()
pareto_a.bar(dados_a.index, dados_a['Valores'], color=cor_barra)
pareto_a.set(xlabel='Países',
           ylabel='Qantidade',
           title='Numero de medalhas obtidas por atletas de 5 países\nnos Jogos Olimpicos de Londres, em 2012')

barras_a = pareto_a.patches
valores_a = dados_a['Valores']
for barra_a, valor_a in zip(barras_a, valores_a):
    altura_a = barra_a.get_height()
    pareto_a.text(barra_a.get_x()+barra_a.get_width()/2, altura_a,
                valor_a, ha='center', va='bottom')

linha_a = pareto_a.twinx()
linha_a.plot(dados_a.index, dados_a['PorcAcum'], 
           color=cor_linha, marker='o', ms=tam_linha)
linha_a.yaxis.set_major_formatter(PercentFormatter())
linha_a.tick_params(axis='y', colors = cor_linha)
```
